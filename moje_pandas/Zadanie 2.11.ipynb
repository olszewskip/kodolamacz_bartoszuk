# E. jaki jest najlepszy film wszechczasów, (najlepszy, czyli ma najwyższą średnią ocenę) - to zadanie możesz rozwiązać wykonując złączenie (join) zbioru movies i ratings

https://pandas.pydata.org/pandas-docs/stable/merging.html

Typy join:
* inner - najpopularniejszy, każdy wiersz musi mieć swój odpowiednik w drugiej ramce
* left - te z lewej ramki nie muszą mieć swoich odpowiedników w prawej
* right - te z prawej nie muszą mieć odpowiedników w lewej
* outer - każdy wiersz z obu znajduje się przynajmniej raz, nawet gdy nie ma odpowiednika (suma left i right)

```python
ratings.head()
```

## Podejście I

```python
movies_ratings = pd.merge(movies, ratings) # lub pd.merge(movies, ratings, on="MovieID"), ale on sam wykrywa wspólne nazwy kolumn
```

```python
tytul_ocena = movies_ratings.groupby("Title")["Rating"].mean().reset_index()
tytul_ocena[tytul_ocena.Rating == tytul_ocena["Rating"].max()]
```

## Podejście II (wydajniejsze)

```python
oceny_movieID = ratings.groupby("MovieID")["Rating"].mean() #średnie liczymy mając tylko id filmów
oceny_movieID
```

```python
# średnie liczymy mając tylko id filmów
oceny_movieID = ratings.groupby("MovieID")["Rating"].mean()
# tytuły dopisujemy już mając średnie
pd.merge(pd.DataFrame(oceny_movieID), movies, left_index=True, right_on="MovieID")\
        .sort_values(by="Rating", ascending=False)[["Title", "Rating"]]
```

# F. wykonaj poprzedni punkt, odrzucając wcześniej filmy które nie uzyskały wystarczająco dużo głosów (np 100)

## podejście I: obliczyć liczności (tylko), a potem zrobić join z średnimi, żeby dostać tylko filmy z odpowiednią licznością

```python
pd.DataFrame(oceny_movieID).head()
```

```python
pd.DataFrame(oceny_movieID).head()
```

```python
oceny_movieID = ratings.groupby("MovieID")["Rating"].mean()
licznosci_movieID = ratings.groupby("MovieID").size()
licznosci_odfiltrowane = licznosci_movieID[licznosci_movieID>100]
srednie_odfiltrowane = pd.merge(pd.DataFrame(oceny_movieID), 
         pd.DataFrame(licznosci_odfiltrowane),
                                on="MovieID").reset_index()[["MovieID", "Rating"]]
#          left_index=True,
#          right_index=True).reset_index()[["MovieID", "Rating"]] # oceny_movieID, licznosci_odfiltrowane
tytuly_srednie = pd.merge(srednie_odfiltrowane, movies)[["Title", "Rating"]]\
    .sort_values(by="Rating", ascending=False).reset_index(drop=True) # wynik powyzszego i movies
tytuly_srednie["Title"][0] # nic nie ginie
```

## podejście II: zrobić z agg zarówno mean jak i count, a potem wyfiltrować

```python
licznosciSrednie = ratings.groupby("MovieID").agg({"Rating" : ["mean", "size", "count"]})
licznosciSrednie
srednie_odfiltrowane = licznosciSrednie[licznosciSrednie[("Rating", "size")] > 100][("Rating", "mean")].reset_index()
srednie_odfiltrowane.columns = ["MovieID", "Rating"]
tytuly_srednie = pd.merge(srednie_odfiltrowane, movies)[["Title", "Rating"]]\
    .sort_values(by="Rating", ascending=False).reset_index(drop=True) # wynik powyzszego i movies
tytuly_srednie
# tytuly_srednie["Title"][0] # nic nie ginie
```

# G. jaki jest najlepszy film według kobiet i według mężczyzn

```python
# trzeba zacząć od merge ratings <-> users
# potem movies <-> reszta
oceny_plec = pd.merge(users, ratings)[["UserID", "Gender", "MovieID", "Rating"]]
oceny_plec.head()
# srednia dla pary (film, plec)
oceny_plec_srednie = oceny_plec.groupby(["MovieID", "Gender"]).agg({"Rating": ["mean", "count"]})
# odfiltrowac > 100
oceny_plec_srednie = oceny_plec_srednie[oceny_plec_srednie[("Rating", "count")] > 100]
# wyłuskać najlepszy dla mężczyzn i kobiet (sprytnie lub nie)
oceny_plec_srednie = oceny_plec_srednie.reset_index()
oceny_plec_srednie[oceny_plec_srednie.Gender=="M"] # dalej jak zawsze
wszystko = oceny_plec_srednie.sort_values(by=("Rating", "mean"), ascending=False)
wszystko.drop_duplicates(subset=[("Gender", "")])
#     .drop_duplicates(subset=["Gender"])

# dodac tytuly filmow
```

# I. jakie trzy gatunki filmowe są najczęściej oglądane przez kobiety i mężczyzn

```python
#jak w D, ale z merge, rozbiciem itd.

wszystko = pd.merge(pd.merge(movies, ratings), users)
mezczyzni = wszystko[wszystko.Gender == "M"].reset_index(drop=True)
kobiety = wszystko[wszystko.Gender == "F"].reset_index(drop=True)

mezczyzni_gatunki = pd.Series(np.concatenate(mezczyzni.Genres.str.split("|"))).value_counts()
kobiety_gatunki = pd.Series(np.concatenate(kobiety.Genres.str.split("|"))).value_counts()
```
