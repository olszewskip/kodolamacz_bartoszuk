# 3

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", con)
sql_result
```

```python
result = planes.groupby("engine").size()
print(result)
result = result.reset_index()
print(result)
result = result[[0,"engine"]]
result.columns = ["COUNT(*)","engine"]
result
```

```python
sql_result.equals(result)
```

```python
result = planes.engine.value_counts()
print(result)
result = result.reset_index()
print(result)
result = result[["engine","index"]]
result.columns = ["COUNT(*)","engine"]
result = result.sort_values(by="engine")
result = result.reset_index(drop=True)
result
```

```python
sql_result.equals(result)
```

```python
pd.testing.assert_frame_equal(sql_result, result) # Error?
```

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type", con)
sql_result
```

```python
result = planes.groupby(["engine","type"]).size().reset_index()
result.columns = ["engine", "type", "COUNT(*)"]
result = result[["COUNT(*)", "engine", "type"]]
result
```

```python
sql_result.equals(result)
```

```python
result = pd.crosstab(planes.engine, planes.type)
result
```

```python
result = pd.crosstab(planes.engine, planes.type).stack() 
result
```

```python
result = pd.crosstab(planes.engine, planes.type).unstack() 
result
```

```python
sql_result = pd.read_sql_query("SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes GROUP BY engine, manufacturer", con)
sql_result
```

```python
result = planes.groupby(["engine","manufacturer"]).agg({"year": ["min", "mean","max"]}).reset_index()
result.columns = ["engine", "manufacturer", "MIN(year)", "AVG(year)", "MAX(year)"]
result = result[["MIN(year)","AVG(year)", "MAX(year)","engine", "manufacturer"]]
result
```

```python
sql_result.equals(result)
```
