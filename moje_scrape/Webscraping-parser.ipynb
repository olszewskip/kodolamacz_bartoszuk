```python
f = requests.get("https://www.filmweb.pl/film/Pi%C4%99%C4%87dziesi%C4%85t+twarzy+Greya-2015-655761")
soup = BeautifulSoup(f.text, 'html.parser')

# soup.find_all(class_="story")
soup.find(itemprop="director").find("a").contents[0]
soup.find(itemprop="director").find("a").get("title")
```

```python
soup.find(text="scenariusz:").parent.next_sibling()[2].get("title")
```

```python
boxoffice = soup.find(text="boxoffice:").next_element.contents[0] #[2].get("title")
boxoffice = boxoffice.replace(" ","")
boxoffice = boxoffice.replace("$", "")
int(boxoffice)
```

```python
# ocena????
```

```python
ocena = soup.find(itemprop="ratingValue").string
ocena = ocena.replace(",", ".")
float(ocena)
```

```python
#wyszukiwanie filmu po tytule
tytul = "50 twarzy Greya"
p = {
        "q" : tytul
    }
soup = BeautifulSoup(requests.get("http://www.filmweb.pl/films/search", params=p).text, 'html.parser')
soup.find_all(class_="filmPreview__title")[0].previous_element.get("href")
link_wzgledny = soup.find_all(class_="filmPreview__link")[0].get("href")
adres = "http://www.filmweb.pl" + link_wzgledny
adres
# print(soup.prettify())
```
