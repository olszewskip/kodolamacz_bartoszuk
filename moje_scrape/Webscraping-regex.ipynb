```python
spolka = "cdr"

# link = fr"https://stooq.pl/q/?s={spolka}\n"
# print(link)

link = f"https://stooq.pl/q/?s={spolka}"
f = requests.get(link)
re.findall("Kurs.*?<span.*?>(.*?)</span>", f.text)
```

```python
url = "https://stooq.pl/q/?s=cdr"
f = requests.get(url)
p = re.compile(r">Kurs.*?(\d+\.\d{2})<")
wyniki = p.findall(f.text)
wyniki
```

```python
re.findall("Zmiana<.*?([+-]\d+\.\d{2}).*?</span>.*?\(([+-]\d+\.\d{2}%)\)</span>", f.text, re.DOTALL)
```

```python
pola=["Kurs", "Zmiana", "Zmiana 1r", "Data", "Max/min"]

for pole in pola:

    found = re.findall(

            fr"{pole}<br>.*?<span.*?>(.*?)</span>",

            f.text

        )

    print(found)
```
