```python
import time
driver = webdriver.Firefox()
driver.get(r"https://nik.bot.nu/brws.html#list/tag:berserk/")

SCROLL_PAUSE_TIME = 1.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

imageBlocks = driver.find_elements_by_class_name("image-block")
    
driver.close()
len(imageBlocks)
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Firefox()
driver.get(r"https://sostest.mini.pw.edu.pl")
username = driver.find_element_by_id("username")
password = driver.find_element_by_id("password")

username.clear()
username.send_keys("252087")
password.clear()
password.send_keys("Iw252087")
password.submit()
time.sleep(1)

driver.get(r"http://sostest.mini.pw.edu.pl/Student/Scores")
#driver.get(r"http://sostest.mini.pw.edu.pl/Student/Task")
#username2 = driver.find_element_by_id("username")
#print(username2)
#print(driver.page_source)
#trzeba poczekac az sie zaladuje

# EC.presence_of_element_located((By.CSS_SELECTOR, "span.pull-right.text-italic.text-underscore"))

# punkty = driver.find_element_by_css_selector("span.pull-right.text-italic.text-underscore")

#https://seleniumhq.github.io/selenium/docs/api/py/webdriver/selenium.webdriver.common.by.html?highlight=locator
punkty = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "span.pull-right.text-italic.text-underscore"))
    )

print(punkty)
print(punkty.text)
driver.close()
```
