# Grafika wysokopoziomowa

```python
p = [0.3,         0.25,       0.4,           0.05]
e = ["Wcale nie", "Ależ tak", "Już mówiłem", "Zapłać, to powiem"]
```

```python
x = np.arange(len(p)) # współrzędne środków słupków
s = 0.7             # szerokość każdego słupka
#plt.bar([0,2,3,4], p, s, color=["0.25", "0.5", "0.75", "1"])
# plt.bar(x, p, s, color=["r", "g", "b", "k"])
plt.bar(x, p, s, color=["r", "g"])
plt.xticks(x, e, rotation=17)      # nowa oś Ox
plt.show()
```

```python
x = np.arange(len(p)) # współrzędne środków słupków
s = 1.5             # szerokość każdego słupka
#plt.bar([0,2,3,4], p, s, color=["0.25", "0.5", "0.75", "1"])
# plt.bar(x, p, s, color=["r", "g", "b", "k"])
plt.bar(x, p, s, color=["#00FF0055", "#FF00FF44"])
plt.xticks(x, e, rotation=17)      # nowa oś Ox
plt.show()
```

```python
import matplotlib.mlab as mlab
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 20, normed=1, facecolor='green', alpha=0.75, edgecolor="k") #normed odpowiada czy procent czy liczba osob

# y = mlab.normpdf(bins, mu, sigma)
# l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
