```python
t = np.linspace(0, 2*np.pi, 100)
r = 1
x = r*np.cos(t)
y = r*np.sin(t)

plt.figure(figsize=(8,8))
plt.fill(x,y, edgecolor="red", facecolor="#FF000040", linewidth=5.0)
# plt.plot(0,0, "ko")
plt.scatter(0,0, color="k")
plt.text(0,0,"(0,0)", 
         horizontalalignment='left',
         verticalalignment='top',
         fontsize=15)
plt.axis("equal")

teraz = datetime.datetime.now()
obrot = ((teraz.hour % 12)/12) * 2*np.pi
kat = np.pi/2-obrot
x_godzinowa = 0.7*np.cos(kat)
y_godzinowa = 0.7*np.sin(kat)
plt.plot([0,x_godzinowa], [0, y_godzinowa])

obrot = ((teraz.minute)/60) * 2*np.pi
kat = np.pi/2-obrot
x_godzinowa = 0.85*np.cos(kat)
y_godzinowa = 0.85*np.sin(kat)
plt.plot([0,x_godzinowa], [0, y_godzinowa])

obrot = ((teraz.second)/60) * 2*np.pi
kat = np.pi/2-obrot
x_godzinowa = 0.95*np.cos(kat)
y_godzinowa = 0.95*np.sin(kat)
plt.plot([0,x_godzinowa], [0, y_godzinowa])

plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)

plt.figure(figsize=(8,8))
plt.fill(x,y, edgecolor="red", facecolor="#00FF0040", linewidth=5.0)
plt.plot(0,0,'ko')
plt.axis("equal")

plt.plot([0,0],[0.75,1], "b")
plt.plot([0,0],[-0.75,-1], "b")
plt.plot([0.75,1],[0,0], "b")
plt.plot([-0.75,-1], [0,0], "b")

plt.text(0,0.75,"12", #'middle',
        horizontalalignment='center',
        verticalalignment='top',
        fontsize=15)
plt.text(0,-0.75,"6", #'middle',
        horizontalalignment='center',
        verticalalignment='bottom',
        fontsize=15)
plt.text(0.75,0,"3", #'middle',
        horizontalalignment='right',
        verticalalignment='center',
        fontsize=15)
plt.text(-0.75,0,"9", #'middle',
        horizontalalignment='left',
        verticalalignment='center',
        fontsize=15)

teraz = datetime.datetime.now()
godzina = (teraz.hour % 12) / 12
godzina_obrot = (0.5 * np.pi) - godzina * (2 * np.pi) - (teraz.minute / 60) * (2*np.pi/12)
x_godzina = np.cos(godzina_obrot)
y_godzina = np.sin(godzina_obrot)
plt.plot([0,0.5*x_godzina],[0,0.5*y_godzina])

minuta = teraz.minute / 60
minuta_obrot = (0.5 * np.pi) - minuta * (2 * np.pi)
x_minuta = np.cos(minuta_obrot)
y_minuta = np.sin(minuta_obrot)
plt.plot([0,0.75*x_minuta],[0,0.75*y_minuta])

sekunda = teraz.second / 60
sekunda_obrot = (0.5 * np.pi) - sekunda * (2 * np.pi)
x_sekunda = np.cos(sekunda_obrot)
y_sekunda = np.sin(sekunda_obrot)
plt.plot([0,0.8*x_sekunda],[0,0.8*y_sekunda])

plt.axis("off")
plt.show()
```

# Mapy kolor√≥w

```python
# https://matplotlib.org/examples/color/colormaps_reference.html
jakaMapa = "plasma"
cmap = matplotlib.cm.get_cmap(jakaMapa)
n = 800
for i in np.arange(n+1):
    plt.fill([i,i+1,i+1,i], [0,0,1,1], color=cmap(i/n))
# plt.axis("equal")
plt.show()
```
