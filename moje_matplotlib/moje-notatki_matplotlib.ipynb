```python
def plot_rainbow(n):
    x = np.array([0, 0, 1, 1])/n
    y = np.array([0, 1, 1, 0])
    def plot_square(i):
        plt.fill(x + np.array([1,1,1,1]) * i/n, y, color=cmap(i/n))
        return None

    for i in range(n):
        plot_square(i)
    
    plt.show()

plot_rainbow(100)
```

```python
tips = sns.load_dataset("tips")
smoker = tips.smoker.value_counts()
smoker
```

```python
plt.pie(smoker, 
        explode = [0.2,0.2],
        labels=smoker.index,
        colors=["r", "b"],
        shadow=True,
        startangle=66)
plt.axis("equal")
plt.legend()
plt.title("Palacze")
leg=plt.legend(title="Legenda", frameon = True)

plt.show()
```

```python
x = np.arange(len(smoker.index))
s = 0.7
plt.bar(x, smoker.values, s, color = ["r", "b"])
plt.grid(False)
plt.legend(smoker.index)
plt.xticks(x, smoker.index, rotation=17)

plt.show()
```
