## podejście 2, starsze, ale działa

```python
import unicodedata
def strip_accents(s):
    wynik = ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
    wynik = wynik.replace("ł", "l")
    wynik = wynik.replace("Ł", "L")
    return wynik
strip_accents("Łłóżźćą")
```

```python
kolumna = oceny.Nazwisko.str.lower() + oceny.Imię.str.lower().str[0] + "@student.mini.pw.edu.pl"
kolumna.apply(strip_accents)
```

```python
oceny2 = oceny.copy()
indeksyDoUsuniecia = []


for i in range(2,16):
    numerPunkty = re.findall(r"Z(\d+)\((\d+)\+(\d+)\)",oceny.columns[i])
    if len(numerPunkty) == 1:
        rozbitaKolumna = oceny.iloc[:,i].str.split("+", expand=True)
        oceny2["L"+numerPunkty[0][0]+"("+numerPunkty[0][1]+")"] = rozbitaKolumna.iloc[:,0]
        oceny2["D"+numerPunkty[0][0]+"("+numerPunkty[0][2]+")"] = rozbitaKolumna.iloc[:,1]
        indeksyDoUsuniecia.append(i)
#         oceny2 = oceny2.drop(oceny.columns[i], axis=1)
    else:
        numerPunkty = re.findall(r"Z(\d+)\((\d+)\)",oceny.columns[i])
        if len(numerPunkty) == 1:
            rozbitaKolumna = oceny.iloc[:,i].str.split("+", expand=True)
            oceny2["L"+numerPunkty[0][0]+"("+numerPunkty[0][1]+")"] = rozbitaKolumna.iloc[:,0]
            indeksyDoUsuniecia.append(i)
#             oceny2 = oceny2.drop(oceny.columns[i], axis=1)
oceny2.drop(columns = oceny2.columns[indeksyDoUsuniecia], inplace=True)
oceny2.drop(columns = "Z15(5)", inplace=True)
oceny2
```

```python
oceny2 = oceny2.replace(["NA","NB", "NS\.\(.+\)"], "0", regex=True)
oceny2 = oceny2.replace(["ZW","US",],np.nan)
oceny2
```

```python
oceny2 = pd.concat([oceny2.iloc[:,0:3],oceny2.iloc[:,3:30].astype(float)], axis=1)
oceny2
```

```python
for columnIndex in range(3, 30):
    for rowIndex in range(oceny.shape[0]):
        if np.isnan(oceny2.iloc[rowIndex, columnIndex]):
            oceny2.iloc[rowIndex, columnIndex] = \
                np.mean(np.r_[oceny2.iloc[:, columnIndex].mean(),\
                              oceny2.iloc[rowIndex, 3:30].mean()])  
oceny2
```

```python
# suma wierszy
oceny2["Suma"] = oceny2.iloc[:,3:30].sum(axis=1)
# suma calosci (znowu regex)
suma = 0
for kolumna in np.r_[3:30]:
    suma += int(re.findall(r"[LD]\d{1,2}\((\d)\)" ,oceny2.columns[kolumna])[0])
suma
# procentowy sukces
oceny2["Procent"] = oceny2.iloc[:,3:30].sum(axis=1)/suma
# pd.cut()
oceny2
```
